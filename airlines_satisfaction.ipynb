```python
#Preprocessing the data 



def encode_gender(x):
    if x=="Male":
        return 0
    elif x=="Female":
        return 1
    else:
        return -1
    
def encode_satisfaction(x):
    if x=="satisfied":
        return 1
    elif x=="neutral or dissatisfied":
        return 0
    else:
        return -1
    
def encode_customerType(x):
    if x=="Loyal Customer":
        return 1
    elif x=="disloyal Customer":
        return 0
    else:
        return -1

def encode_travelType(x):
    if x=="Personal Travel":
        return 0
    elif x=="Business travel":
        return 1
    else:
        return -1
    
def encode_class(x):
    if x=='Business':
        return 2
    elif x=="Eco Plus":
        return 1
    elif x=="Eco":
        return 0
    else:
        return -1
    
def preprocess(df):
    df=df.drop(['Unnamed: 0','id'], axis=1)
    df['Gender']=df['Gender'].apply(encode_gender)
    df['satisfaction']=df['satisfaction'].apply(encode_satisfaction)
    df['Customer Type']=df['Customer Type'].apply(encode_customerType)
    df['Type of Travel']=df['Type of Travel'].apply(encode_travelType)
    df['Class']=df['Class'].apply(encode_class)
    df['Arrival Delay in Minutes'].fillna(df['Arrival Delay in Minutes'].median(), inplace = True)
    return df
    
train=preprocess(train_data)
test=preprocess(test_data)
```

```python
#Train and test data split

X_train=train.drop(['satisfaction'], axis=1)
y_train=train['satisfaction']

X_test=test.drop(['satisfaction'], axis=1)
y_test=test['satisfaction']
```

```python
# Normalize Features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
#Building classifiers-Random Forest
from sklearn.ensemble import RandomForestClassifier
randomforestclf=RandomForestClassifier(n_estimators=1000,criterion='entropy', max_depth=25,
         min_samples_leaf= 1,
         min_samples_split= 2,random_state=42,n_jobs=-1)
randomforestclf.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score 
print("Cross Validation Scores: ", cross_val_score(randomforestclf, X_train,y_train))
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,roc_auc_score, plot_confusion_matrix,plot_roc_curve
import matplotlib.pyplot as plt

y_pred = randomforestclf.predict(X_test)

print(classification_report(y_test,y_pred))
print('\n')
print('Confusion matrix : \n{}'.format(confusion_matrix(y_test,y_pred)))
print('\n')
print('Accuracy score : {}'.format(accuracy_score(y_test,y_pred)))
roc_auc_rf = roc_auc_score(y_test, y_pred)
print("ROC_AUC = {}".format(roc_auc_rf))
plot_confusion_matrix(randomforestclf, X_test, y_test,cmap=plt.cm.Blues, normalize = 'all')
plot_roc_curve(randomforestclf, X_test, y_test)

acc_random_forest = accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV

rfc=RandomForestClassifier(random_state=42)
param_grid = { 
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}
scores=['precision','recall']
for score in scores:
    
    print("# Tuning hyper-parameters for %s" % score)
    print()
    
    clf=GridSearchCV(estimator=rfc, param_grid=param_grid, scoring='%s_macro' % score, cv=3 )
    clf.fit(X_train, y_train)
    
    print("Best parameters set found on development set:")
    print()
    print(clf.best_params_)
    print()
    print("Grid scores on development set:")
    print()
    means = clf.cv_results_['mean_test_score']
    stds = clf.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, clf.cv_results_['params']):
        print("%0.3f (+/-%0.03f) for %r"
              % (mean, std * 2, params))
    print()

    print("Detailed classification report:")
    print()
    print("The model is trained on the full development set.")
    print("The scores are computed on the full evaluation set.")
    print()
    y_true, y_pred = y_test, clf.predict(X_test)
    print(classification_report(y_true, y_pred))
    print()
```
